```python
from modules import courses
import pydot
from IPython.display import Image
```

```python
G = pydot.Dot(graph_type='digraph')
```

```python
node = pydot.Node('First', fillcolor='red')
node2 = pydot.Node('Second', color='blue')
node3 = pydot.Node('Third', color='green')
edge = pydot.Edge('First', 'Second', color='red')
```

```python
G.add_node(node)
G.add_node(node2)
G.add_node(node3)
G.add_edge(edge)

Image(G.create_png())
```
