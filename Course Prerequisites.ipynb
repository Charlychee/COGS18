# Course Prerequisite Finder
When planning courses to take in college, many courses have prerequisite courses which must be taken beforehand. It is left up to the student to plan out their courses. This can leave a student overwhelmed at which courses they must take in which order since, very often, prerequisite courses have their own prerequisite courses. This can become a disaster very quickly. The Course Prerequisite Finder tool aids in finding all the courses that a student must take before they can take a desired course. It helps the student visually see how the courses relate to one another and can make more informed decisions while planning their college course schedule. <br /><br />
Follow the arrows to find which courses you must take as a prerequisite. <br />
**Red** arrows represent courses which all must be completed. <br />
**Blue** arrows represent a cluster of courses which you only need to complete one of. Different shades of blue represent different clusters.<br />
**Green** arrows represent courses which can be taken at the same time or has already been completed.

```python
from modules import courses
from IPython.display import Image
```

In the following cell, we are looking for the prerequisites for the course, CSE100. After the cell is run, we can see that to take CSE100, the student must have completed CSE15L and CSE21, (CSE30 or CSE5A or ECE15 or MAE9), and (CSE21 or MATH154 or MATH184A) We can also see that CSE15L and CSE12 must be taken as concurrent courses. Each of those courses may have their own prerequisites, which can be followed down the chain of arrows.

```python
course = 'CSE100'
graph = courses.get_prereqs(course)
Image(graph.create_png())
```

# Try it out yourself!

This tool uses the data from the UCSD_courses.json file to find the prerequisite courses. If you cannot find your courses and their prerequisites in the json file, please add them in before using this tool for that course. They should be formatted as the following.<br /><br />
{<br />
&nbsp;&nbsp;&nbsp;"COURSE_ID" : ["and1", "and2", "and3", "conc*", ["or1", "or2", "or3"]]<br />
}<br />
- **and** - Courses that must be taken
- **or** - Courses where only one of the courses in the list needs to be taken
- **conc** - Courses that can be taken at the same time as the desired course
